```python
import time
# get 'lead_paragraph', 'headline': {'main'}, 'pub_date'
nyt_key = 'MvMqIkjspz2Ljj6CkJFQtwYnFjtzdQK6'

def get_content(yr_from, yr_to, previous_set=set(), mth_from=1, mth_to=13):
    '''
    Get article content from the nyt's API
    '''
    article_lst = []
    id_set = previous_set
    errors = []
    # loop through years of the decade
    for year in range(yr_from, yr_to + 1):
        for month in range(mth_from, mth_to):
            if month > 1:
                time.sleep(6) # wait 6 seconds between each call
            try:
                call = 'https://api.nytimes.com/svc/archive/v1/{}/{}.json?api-key={}'.format(year, month, nyt_key)
                response = requests.get(call)
                json_response = response.json()
                print('working on {} / {}'.format(year, month))
                articles = json_response['response']['docs']
                for _, article in enumerate(articles):
                    try:
                        # keep id, date, headline, lead_paragraph
                        id_ = article['_id']
                        if id_ not in id_set:
                            id_set.add(id_)
                            date = article['pub_date']
                            headline = article['headline']['main']
                            #lead_paragraph = article['lead_paragraph'] , lead_paragraph
                            article_lst.append([id_, date, headline])
                    except Exception as e:
                        # article error
                        #print('error in json:', e)
                        errors.append((id_, article))
                        continue
            except Exception as e:
                # call error
                print('error in call:', e)
                continue

    return article_lst, id_set, errors
```

```python
def get_decades(decades):
    '''
    Get content by decade into csv file
    '''
    results_ = []
    for decade in decades:
        decade_start = int(decade[: 4])
        decade_end = int(decade[5: ])
        print('=> working on decade {}-{}'.format(decade_start, decade_end))
        lst_, set_, errors_ = get_content(decade_start, decade_end)
        with open(decade + '.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(lst_)
            print('==> saved csv')
        results_.append([decade, set_, errors_])
    return results_      
```
